```python
def ввод_координат():                               # назначаем функцию для ввода Х и О
    
    while True:                                     # устанавливаем цикл пока выполняется условие
        координаты = input("Ваш ход: ").split()     # вводим координаты Х или О, с помощью split разбиваем на отдельные элементы, если установить ограничение int не работает следующая итерация цикла
        
        if len(координаты) != 2:                    # делаем проверку на правильность ввода, если длина полученного списке не равна 2
            print('Введите 2 координаты клетки через пробел!')   # выводим сообщение
            continue                                # и переходим на следующую итерацию while, значит снова запрашиваем координаты
       
        x, y = координаты                           # производим распаковку значений
        
        if not(x.isdigit()) or not(y.isdigit()):    # осуществляем проверку на то, что введенные значения это числа, если нет 
            print('Введите числа')                  # выводим сообщение
            continue                                # и переходим на следующую итерацию
        
        x, y = int(x), int(y)                      # после проверки что координат 2 и это числа, присваиваем распакованным переменным тип целого числа, без этого выходит ошибка "'>' not supported between instances of 'int' and 'str'"
        
        if 0 > x or x > 2 or 0 > y or y > 2:        # устанавливаем ограничения для координат: если меньше 0 и больше 2
            print('Ход вне поля')                   # выводим сообщение
            continue                                # и переходим на следующую итерацию
            
        if поле[x][y] != "   ":                     # если введенные координаты не совпадают с пустой клеткой
            print('Клетка занята!')                 # выводим сообщение
            continue                                # и переходим на следующую итерацию
            
        return x, y                                 # координаты принимаются при условии, что прошли все проверки
ввод_координат()                                    #  если не запустить функцию в строке, потом она не включается
```

```python
def проверка():                                              # запускаем функцию проверки с универсальным описанием комбинаций   
        for i in range(3):                                   # запускаем цикл проверки комбинаций на выигрышные строки
            символ = []                                      # назначаем переменную со списком координат
            for j in range(3):                               # внутри цикла строки запускаем цикл проверки столбца
                символ.append(поле[i][j])                    # добавляем по строке 1-3 столбец 1-3
            if символ == [" X "," X "," X "]:                # если в списке символов строка из 3х Х
                print('Выиграл Х!')                         # выводим сообщение
                return True                                       
            elif символ == [" 0 "," 0 "," 0 "]:              # если в списке строка столбец из 3х 0
                print('Выиграл 0!')                         # выводим сообщение
                return True                                            
                                                     
         
        for i in range(3):                                   # запускаем цикл проверки комбинаций на выигрышные столбцы
            символ = []                                      # назначаем переменную со списком координат
            for j in range(3):                               # внутри цикла строки запускаем цикл проверки столбца
                символ.append(поле[j][i])                    # добавляем по столбцу 1-3 строку 1-3
            if символ == [" X "," X "," X "]:                # если в списке символов столбец из 3х Х
                print('Выиграл Х!')                         # выводим сообщение
                return True                                       
            elif символ == [" 0 "," 0 "," 0 "]:              # если в списке символов столбец из 3х 0
                print('Выиграл 0!')                         # выводим сообщение
                return True                                            
                                               
              
        символ = []                                          # назначаем переменную со списком координат для проверки диагонали
        for i in range(3):                                   # запускаем цикл проверки строки с индексами 1-3
            символ.append(поле[i][i])                        # добавляем по сироке 1-3 строку с индексом столбца = индекс строки
        if символ == [" X "," X "," X "]:                    # если в списке символов диагональ из 3х Х
            print('Выиграл Х!')                             # выводим сообщение
            return True
        elif символ == [" 0 "," 0 "," 0 "]:                  # если в списке символов диагональ из 3х 0
            print('Выиграл 0!')                             # выводим сообщение
            return True                                            
                                                    
  
        символ = []                                          # назначаем переменную со списком координат для проверки диагонали
        for i in range(3):                                   # запускаем цикл проверки строки
            символ.append(поле[i][2 - i])                    # добавляем по строке 1-3 столбец с индексом = (2-индекс строки)
        if символ == [" X "," X "," X "]:                    # если в списке символов диагональ из 3х Х
            print('Выиграл Х!')                             # выводим сообщение
            return True
        elif символ == [" 0 "," 0 "," 0 "]:                  # если в списке символов диагональ из 3х 0
            print('Выиграл 0!')                             # выводим сообщение
            return True                                            
        return False    
    
проверка()
```

```python
""Игра "Крестики - нолики".""

приветствие()                           # 1. выводим функцию приветствия

поле = [["   "] * 3  for i in range(3)] # 2. устанавливаем внешнюю переменную для объявления игрового поля в виде 2х мерной мартицы 3х3 с пустыми пробелами

игровое_поле()                          # 3. выводим функцию отображения пустого игрового поля для наглядности

номер_хода = 0                          # 4. Заводим игровой цикл: 4.1 заводим счетчик для ходов игроков
while True:                             # 4.2 задаем бесконечный цикл, пока правда
    номер_хода += 1                       # ход + 1
    
    игровое_поле()                        # 4.3. внутри игрового цикла запускаем функцию отображения игрового поля для вывода со значениями игроков
    
    if номер_хода % 2 == 1:               # 4.4 задаем условие для опредения какой игрок ходит, т.к. ходы поочередно, то задаем, что крестики всегда четные ходы
        print('Ходит Х')                    # выводим сообщение
    else:                                 # если ход не четный
        print('Ходит 0')                    # выводим сообщение
        
    x, y = ввод_координат()               # 4.5 внутри игрового цикла вводим функцию ввода координат хода игроками
    
    if номер_хода % 2 == 1:               # 4.6 снова запускаем условие определения хода
        поле[x][y] = ' X '                 # теперь если условие выполнено на поле отображается Х, для обозначения вывода берем в кавычки, что это не переменная а элемент отображения
    else:                                  # если условие не выполнено 
        поле[x][y] = ' 0 '                 # на поле отображается 0
        
    if проверка():                        # 4.7 внутри игровго цикла вводим функцию проверки выигрышных ходов
        break                             # если проверка True, то игра завершена, выводится сообщение о выигрывшем из функции
    
    if номер_хода == 9:                   # 4.8 внутри игрового цикла вводим проверку на ничью, если максимальное количество ходов достигнуто.!!!через оператор присваивания, если установить = будет ругаться на синтаксис
        print('Ничья')                     # выводим сообщение
        break                              # добавляем условия выхода из бесконечного цикла прекратить цикл
        
```
